```python
import os
import shutil
from sklearn.model_selection import train_test_split
```

```python
# Dataset dizininin yolu
dataset_dir = 'mantar_veri_seti'

# Yeni dizinler
directories = ['train', 'validation', 'test']

# Zehirli ve Zehirsiz için alt klasörler
subfolders = ['zehirli', 'zehirsiz']

# Dizini oluşturma
for dir in directories:
    for subfolder in subfolders:
        os.makedirs(os.path.join(dataset_dir, dir, subfolder), exist_ok=True)
```

```python
# Görsel dizinlerinin yolları
zehirli_dir = os.path.join(dataset_dir, 'zehirli')
zehirsiz_dir = os.path.join(dataset_dir, 'zehirsiz')

# Zehirli ve Zehirsiz görselleri listeleyelim
zehirli_images = [os.path.join(zehirli_dir, f) for f in os.listdir(zehirli_dir)]
zehirsiz_images = [os.path.join(zehirsiz_dir, f) for f in os.listdir(zehirsiz_dir)]

# Görselleri train/test olarak ayıralım
zehirli_train, zehirli_test = train_test_split(zehirli_images, test_size=0.15, random_state=42)
zehirsiz_train, zehirsiz_test = train_test_split(zehirsiz_images, test_size=0.15, random_state=42)

# Test ve Validation verisini ayıralım
zehirli_train, zehirli_val = train_test_split(zehirli_train, test_size=0.15, random_state=42)
zehirsiz_train, zehirsiz_val = train_test_split(zehirsiz_train, test_size=0.15, random_state=42)

# Train setini, validation ve test için ayırdık
```

```python
def move_images(image_list, target_folder):
    for image in image_list:
        shutil.copy(image, target_folder)

# Zehirli görselleri yerleştirme
move_images(zehirli_train, os.path.join(dataset_dir, 'train', 'zehirli'))
move_images(zehirsiz_train, os.path.join(dataset_dir, 'train', 'zehirsiz'))
move_images(zehirli_val, os.path.join(dataset_dir, 'validation', 'zehirli'))
move_images(zehirsiz_val, os.path.join(dataset_dir, 'validation', 'zehirsiz'))
move_images(zehirli_test, os.path.join(dataset_dir, 'test', 'zehirli'))
move_images(zehirsiz_test, os.path.join(dataset_dir, 'test', 'zehirsiz'))
```

normalizasyon

```python
from PIL import Image
import numpy as np
import os

def resize_and_normalize(image_path, target_size=(224, 224)):
    # Görseli açma
    img = Image.open(image_path)
    
    # Görseli RGB formatında açma (Eğer başka formatta ise)
    img = img.convert('RGB')
    
    # Görseli yeniden boyutlandırma
    img = img.resize(target_size)
    
    # Görseli numpy array'e çevirme
    img_array = np.array(img)
    
    # Normalizasyon (0-1 arası)
    img_array = img_array / 255.0
    
    return img_array
```

```python
import os
from PIL import Image

def process_and_save_images(image_list, target_folder, target_size=(224, 224)):
    for image_path in image_list:
        # Görseli boyutlandırma ve normalize etme
        processed_image = resize_and_normalize(image_path, target_size)
        
        # Görselin yeni yolunu oluştur
        filename = os.path.basename(image_path)
        new_image_path = os.path.join(target_folder, filename)
        
        # Görseli kaydet
        Image.fromarray((processed_image * 255).astype(np.uint8)).save(new_image_path)
```

```python
import os

# Klasörlerin yolu
dataset_dir = 'mantar_veri_seti'  # Dataset klasörünüzün yolu

# Train, Validation, Test klasörlerine ait dizinler
train_dir = os.path.join(dataset_dir, 'train')
validation_dir = os.path.join(dataset_dir, 'validation')
test_dir = os.path.join(dataset_dir, 'test')

# Zehirli ve Zehirsiz alt klasörlerinin yolları
train_zehirli_dir = os.path.join(train_dir, 'zehirli')
train_zehirsiz_dir = os.path.join(train_dir, 'zehirsiz')
validation_zehirli_dir = os.path.join(validation_dir, 'zehirli')
validation_zehirsiz_dir = os.path.join(validation_dir, 'zehirsiz')
test_zehirli_dir = os.path.join(test_dir, 'zehirli')
test_zehirsiz_dir = os.path.join(test_dir, 'zehirsiz')

# Klasördeki görselleri listeleyelim
train_zehirli_images = [os.path.join(train_zehirli_dir, f) for f in os.listdir(train_zehirli_dir)]
train_zehirsiz_images = [os.path.join(train_zehirsiz_dir, f) for f in os.listdir(train_zehirsiz_dir)]

validation_zehirli_images = [os.path.join(validation_zehirli_dir, f) for f in os.listdir(validation_zehirli_dir)]
validation_zehirsiz_images = [os.path.join(validation_zehirsiz_dir, f) for f in os.listdir(validation_zehirsiz_dir)]

test_zehirli_images = [os.path.join(test_zehirli_dir, f) for f in os.listdir(test_zehirli_dir)]
test_zehirsiz_images = [os.path.join(test_zehirsiz_dir, f) for f in os.listdir(test_zehirsiz_dir)]

# Görselleri işleyip kaydedelim
process_and_save_images(train_zehirli_images, train_zehirli_dir)
process_and_save_images(train_zehirsiz_images, train_zehirsiz_dir)

process_and_save_images(validation_zehirli_images, validation_zehirli_dir)
process_and_save_images(validation_zehirsiz_images, validation_zehirsiz_dir)

process_and_save_images(test_zehirli_images, test_zehirli_dir)
process_and_save_images(test_zehirsiz_images, test_zehirsiz_dir)
```

normalizasyonu kontrol etme (istenen görseller dogru boyutta mı ? )

```python
import os
from PIL import Image

def check_image_sizes(directory, target_size=(224, 224)):
    # Belirtilen dizindeki tüm görselleri listele
    image_files = [os.path.join(directory, f) for f in os.listdir(directory) if f.endswith(('jpg', 'jpeg', 'png'))]
    
    # Her görselin boyutunu kontrol et
    for image_path in image_files:
        img = Image.open(image_path)
        if img.size != target_size:
            print(f"Görselin boyutu uyumsuz: {image_path} - Boyut: {img.size}")
        else:
            print(f"Görsel boyutu uyumlu: {image_path} - Boyut: {img.size}")

# Örneğin, 'train/zehirli' klasöründeki görsellerin boyutlarını kontrol edelim
check_image_sizes('mantar_veri_seti/train/zehirli')
check_image_sizes('mantar_veri_seti/train/zehirsiz')
check_image_sizes('mantar_veri_seti/validation/zehirli')
check_image_sizes('mantar_veri_seti/validation/zehirsiz')
check_image_sizes('mantar_veri_seti/test/zehirli')
check_image_sizes('mantar_veri_seti/test/zehirsiz')
```

modelleme ortak ön hazırlık

```python
import os
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping

# Veri yolları
base_dir = "mantar_veri_seti"
train_dir = os.path.join(base_dir, "train")
val_dir = os.path.join(base_dir, "validation")
test_dir = os.path.join(base_dir, "test")

# Görselleri ön işleme: yeniden boyutlandırma + normalize
image_size = (224, 224)
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255)
val_datagen   = ImageDataGenerator(rescale=1./255)
test_datagen  = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode="binary"
)

val_generator = val_datagen.flow_from_directory(
    val_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode="binary"
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=image_size,
    batch_size=batch_size,
    class_mode="binary"
)
```

VGG16 TABANLI MODEL

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.models import Model

# Temel model
vgg_base = VGG16(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
vgg_base.trainable = False

x = Flatten()(vgg_base.output)
x = Dense(128, activation='relu')(x)
x = Dense(1, activation='sigmoid')(x)

vgg_model = Model(inputs=vgg_base.input, outputs=x)
vgg_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Eğitim
vgg_history = vgg_model.fit(
    train_generator,
    epochs=30,
    validation_data=val_generator,
    callbacks=[EarlyStopping(patience=3, restore_best_weights=True)]
)
```

 MobileNetV2 Tabanlı Model

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import GlobalAveragePooling2D

# Temel model
mobilenet_base = MobileNetV2(weights='imagenet', include_top=False, input_shape=(224, 224, 3))
mobilenet_base.trainable = False

x = GlobalAveragePooling2D()(mobilenet_base.output)
x = Dense(128, activation='relu')(x)
x = Dense(1, activation='sigmoid')(x)

mobilenet_model = Model(inputs=mobilenet_base.input, outputs=x)
mobilenet_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Eğitim
mobilenet_history = mobilenet_model.fit(
    train_generator,
    epochs=20,
    validation_data=val_generator,
    callbacks=[EarlyStopping(patience=3, restore_best_weights=True)]
)
```

ORTAK TEST AŞAMASI

birkaç model daha denendi ama buraya eklemedim aralarından en iyi sonucu MobileNetV2 verdi o yüzden onun üzerinden oynamalar yapıldı .

```python
# VGG16 Test
vgg_loss, vgg_acc = vgg_model.evaluate(test_generator)
print(f"VGG16 Test Accuracy: {vgg_acc:.4f}")

# MobileNetV2 Test
mobilenet_loss, mobilenet_acc = mobilenet_model.evaluate(test_generator)
print(f"MobileNetV2 Test Accuracy: {mobilenet_acc:.4f}")
```
